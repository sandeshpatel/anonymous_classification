```python

#anova test
cat = ['f1', 'f2', 'f3']
def anova(frame):
    anv = pd.DataFrame()
    anv['features'] = cat
    pvals = []
    for c in cat:
           samples = []
           for cls in dtrain[c].unique():
                  s = dtrain[dtrain[c] == cls]['y'].values
                  samples.append(s)
           pval = stats.f_oneway(*samples)[1]
           pvals.append(pval)
    anv['pval'] = pvals
    return anv.sort_values('pval')


k = anova(dtrain.y.values) 
k['disparity'] = np.log(1./k['pval'].values) 
sns.barplot(data=k, x = 'features', y='disparity') 
plt.xticks(rotation=90) 
plt
```

it is clear that even hash value of string is a very good feature

### Feature Engineering

feature f1 can be replaced by the probabilty of each class for a given value of feature f1

lets split out dtrain into two parts names dtrain and d_evaluation. d_evaluation will not be used in model preparation and will be used to evaluate the model at last

```python
dtrain, d_evaluation = train_test_split(dtrain, test_size=0.2)
```

```python
dtrain.shape
```

```python
#one hot encoding of feature f1
lbe = preprocessing.LabelEncoder()
lbe.fit(dtrain.f1.append(dtest.f1).append(d_evaluation.f1))
labeled_f1 = lbe.transform(dtrain.f1).reshape(-1,1)
enc = OneHotEncoder(n_values=264, sparse=False)
enc.fit(labeled_f1)
encoded_f1 = enc.transform(labeled_f1)

```

```python
encoded_f1[0]
```

```python
f1_train, f1_test, y_train, y_test = train_test_split(encoded_f1, dtrain.y, test_size=0.3)
```

```python
clf = BernoulliNB(alpha=0.01)
clf.fit(f1_train, y_train)
y_pred = clf.predict(f1_test)
```

```python

print(accuracy_score(y_test, y_pred))
```

```python
#10 fold cross validation 
alfa_list = np.array([0.0001, 0.0003,0.0009, 0.001,0.003,0.009,0.01, 0.03, 0.09, 0.1, 0.3, 0.9, 1, 3,9,27])
cross_val_scores = []
for alfa in alfa_list:
    classifier = BernoulliNB(alpha=alfa)
    val_score = cross_val_score(classifier, f1_train, y_train, cv=5, scoring='accuracy')
    cross_val_scores.append(np.median(val_score))
    
alpha = alfa_list[np.argmax(cross_val_scores)]
print(cross_val_scores)
print(alpha)
```

```python
# y_pred_f1 contains log probabilities and will be used as feature set in final model
clf_f1 = BernoulliNB(alpha=alpha, fit_prior=False)
clf_f1.fit(encoded_f1, dtrain.y)
y_pred_f1 = clf_f1.predict_proba(encoded_f1)
```

```python
#from sklearn.metrics import accuracy_score
#print(accuracy_score(y_test, y_pred_f1))
#accuracy score = 0.62
```

Feature f2 can be replaced by class count for that feature category

```python
#creating a dictionry with category of f2 as key and each class count for that category
class_count_f2 = {}
for i in range(2996):
    class_count_f2['V' + str(i)] = np.zeros(9)
for row in dtrain.values:
    class_count_f2[row[1]][row[3]-1] +=1
    
encoded_f2 = [class_count_f2[cat] for cat in dtrain.f2]
```

class_count_f2.values() will be used as featues in final model


as the hash value of feature f3 itself is a good feature it can be used as a separate feature in final model

```python
class_count_f3 = {}
for row in dtrain.values:
    if row[2] in class_count_f3:
        class_count_f3[row[2]][row[3]-1] += 1
    else:
        class_count_f3[row[2]] = np.zeros(9)
        class_count_f3[row[2]][row[3]-1] += 1
encoded_f3 = [class_count_f3[cat] for cat in dtrain.f3]
```

```python
tfidf_vectorizer = TfidfVectorizer(ngram_range=(1,2))
tfidf_vectorizer.fit(dtrain.f3.values)
text_data = tfidf_vectorizer.transform(dtrain.f3.values)
```

```python
f3_train, f3_test, y_train, y_test = train_test_split(text_data, dtrain.y, test_size=0.3)
```

```python
#10 fold cross validation 
alfa_list = np.array([0.0001, 0.0003,0.0009, 0.001,0.003,0.009,0.01, 0.03, 0.09, 0.1, 0.3, 0.9, 1, 3,9,27])
cross_val_scores = []
for alfa in alfa_list:
    classifier = MultinomialNB(alpha=alfa)
    val_score = cross_val_score(classifier, f3_train, y_train, cv=5, scoring='accuracy')
    cross_val_scores.append(np.median(val_score))
    
alpha = alfa_list[np.argmax(np.array(cross_val_scores))]
print(cross_val_scores)
print(alpha)
```

```python
clf = MultinomialNB(alpha=alpha)
clf.fit(f3_train, y_train)
y_pred = clf.predict(f3_test)
print(accuracy_score(y_test, y_pred))
```

```python
clf_text = MultinomialNB(alpha=alpha, fit_prior=False)
clf_text.fit(text_data, dtrain.y)
y_pred_f3 = clf_text.predict_proba(text_data)
```

now we have featurized feature f1 in y_pred_f1, feature  f2 in encoded_f2, text data as categorical variable in encoded_f3 and text data in y_pred_f3.

Now we will combine all data to form a final dataframe and the build a classifier over that

```python
df = pd.DataFrame(np.hstack((y_pred_f1, encoded_f2, encoded_f3, y_pred_f3)))
```
